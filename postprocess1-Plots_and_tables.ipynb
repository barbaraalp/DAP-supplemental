```python
def reorder(df):
    new_index = ['F1_Accident','F1_No-Accident','F1_MacroAvg','F1_MicroAvg',
    'F1_WeightedAvg','Precision_Accident','Precision_MacroAvg','Precision_MicroAvg',
    'Precision_No-Accident','Precision_WeightedAvg','Recall_Accident',
    'Recall_MacroAvg','Recall_MicroAvg','Recall_No-Accident','Recall_WeightedAvg']
    return df.reindex(new_index,level=1)
```

```python
def make_df_for_only(city):
    df_1 = pd.read_hdf(city+'.h5',key='LR_only_')
    df0 = pd.read_hdf(city+'.h5',key='GBC_only_')
    df1 = pd.read_hdf(city+'.h5',key='DNN_only_')
    df2 = pd.read_hdf(city+'.h5',key='RNN_only_')
    df3 = pd.read_hdf(city+'.h5',key='RNN_embed_only_')
    #df4 = pd.read_hdf(city+'.h5',key='RNN_embed_att_only')
    #df_1 = reorder(df_1)
    df_1['city'] = city
    df_1 = df_1.reset_index().set_index(['city','index'])
    display(df_1)
    #df0 = reorder(df0)
    df0['city'] = city
    df0 = df0.reset_index().set_index(['city','index'])
    
    #df1 = reorder(df1)
    df1['city'] = city
    df1 = df1.reset_index().set_index(['city','index'])
    
    #df2 = reorder(df2)
    df2['city'] = city
    df2 = df2.reset_index().set_index(['city','index'])
    
    #df3 = reorder(df3)
    df3['city'] = city
    df3 = df3.reset_index().set_index(['city','index'])
   
    
    """
    df4['city'] = city
    df4 = df4.reset_index().set_index(['city','index'])
    """
    
    df_1= pd.concat({"LR": pd.DataFrame(df_1.to_dict(),columns=df_1.to_dict().keys())}, axis=1, names=["Model", "category"],sort=False)
    df_1[' ']="___"
    
    df0= pd.concat({"GBC": pd.DataFrame(df0.to_dict(),columns=df0.to_dict().keys())}, axis=1, names=["Model", "category"],sort=False)
    df0[' ']="___"
    
    df1= pd.concat({"DNN": pd.DataFrame(df1.to_dict(),columns=df1.to_dict().keys())}, axis=1, names=["Model", "category"],sort=False)
    df1[' ']="___"
    df2= pd.concat({"RNN": pd.DataFrame(df2.to_dict(),columns=df2.to_dict().keys())}, axis=1, names=["Model", "category"],sort=False)
    df2[' ']="___ "
    df3= pd.concat({"RNN_embed": pd.DataFrame(df3.to_dict(),columns=df3.to_dict().keys())}, axis=1, names=["Model", "category"],sort=False)
    """
    df3[' ']="__ "
    df4= pd.concat({"RNN_embed_att": pd.DataFrame(df4.to_dict(),columns=df4.to_dict().keys())}, axis=1, names=["Model", "category"])
    """#df3[' ']="__ "
    final = reorder(pd.concat([df_1,df0,df1,df2,df3],axis=1,sort=False))
    
    return final
```
